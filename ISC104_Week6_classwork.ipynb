```python
#p.10
dl = []
for y in range(3): #要素が3個のリストを3個作る
  xl = []
  for x in range(3):
    xl.append(x+y)
  dl.append(xl) #3つのリストを多重リストとしてまとめる
print(dl)
```

```python
#p.11
import matplotlib.pyplot as plt #matplotlibはグラフ描画のためのライブラリ
imglist = [[0,0.25],[0.5,1]] #上の段が1個目のリスト、下の段は2個目のリスト
plt.imshow(imglist,cmap = 'gray') #imshowは画像ファイルや画像情報が格納されたリストを表示する
plt.show()

imglist = [[[1,0,0],[0.5,0,0.5]],[[0,0.5,0.5],[0,0,1]]] #1つのリストが色を指定
plt.imshow(imglist)
plt.show()
```

```python
#p.12
imglist = []
size = 10
for y in range(size): #数値の変化を色のグラデーションで表している（１種のグラフ）
  xlist = []
  for x in range(size):
    val = (x*y)
    xlist.append(val)
  imglist.append(xlist)
plt.imshow(imglist,cmap='gray')
```

```python
#p.13
numlist = []
f = open('numfile-single.csv','r')
for line in f:
  stripped = line.strip()
  numlist.append(float(stripped))
f.close()
print(numlist)
plt.plot(numlist)
plt.show()
```
