```python
#EX8_1 階乗計算の関数を作る
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n - 1)

print(factorial(3))
```

```python
#EX8_2 0!~9!をグラフで描画する
import matplotlib.pyplot as plt

xlist = list(range(10))
ylist = list(map(factorial, xlist))

plt.plot(xlist, ylist)
plt.show()
```

```python
#EX8_3 nCrを計算する関数を作る
import matplotlib.pyplot as plt
def combi(n, r):
  if r == 0 or r == n:
    return 1
  else:
    return combi(n - 1, r) + combi(n - 1, r - 1)

print(combi(5, 3))

#Extra
print(factorial(5) / (factorial(3) * factorial(5 - 3))) #5C2 = 5!/3!*2!になることを確認

#EX8_4 n=20に固定して値の変化をグラフに描画する
n = 20
ylist = []
for r in range(n + 1): #rは0~20の範囲
  ylist.append(combi(n, r))
plt.plot(ylist)
plt.show()

#Extra
ylist2 = list(map(lambda r: combi(n, r), range(n + 1))) #lambda=combi関数に1~n+1を代入
plt.plot(ylist2) #EX8_4と同じグラフが描画された
plt.show()

numlist = [2, 3, 5, 10] #nに適当な数字を割り当てる
for n in numlist:
  total = 0
  for r in range(n + 1):
    combination = combi(n, r)
    total += combination
  print("n=", n, "total=", total) #2^nの関係になっている
```

```python
#EX8_8　ファイル名とファイルサイズを出力する関数を作る
import os
def flatls_size(argpath):
    for entry in os.scandir(argpath):
        if entry.is_file():
            print("File:", entry.path, "Size:", entry.stat().st_size, "bytes")
        elif entry.is_dir():
            print("Dir:", entry.path, "Size:", entry.stat().st_size, "bytes")

flatls_size("../datalab/web")
```

```python
#EX8_9 ディレクトリ内のファイルサイズの合計を返す関数を作る
def flatls_rtsize(argpath):
    total_size = 0
    for entry in os.scandir(argpath):
      file_size = entry.stat().st_size
      total_size += file_size
    return total_size

total_size = flatls_rtsize("../datalab/web")
print("Total size:", total_size, "bytes")
```

```python
#EX8_10　サブディレクトリ内のファイルサイズも含めた合計を返す関数を作る
def recls_size(argpath):
    total_size = 0
    for entry in os.scandir(argpath):
      if entry.is_file():
        file_size = entry.stat().st_size
        total_size += file_size
      elif entry.is_dir():
        total_size += recls_size(entry.path) #再帰関数の利用
    return total_size

total_size = recls_size("../datalab/web")
print("Total size of all files:", total_size, "bytes")
#EX8_9よりもファイルサイズが増えることを確認
```

```python
#EX8_11 ファイルパスをkey、データサイズをvalueとする辞書を返す関数を作る
def recls_sizedict(argpath):
  file_sizedict = {}
  for entry in os.scandir(argpath):
      if entry.is_file():
        file_sizedict[entry.path] = f"{entry.stat().st_size} bytes"
      elif entry.is_dir():
        file_sizedict.update(recls_sizedict(entry.path))
  return file_sizedict

print(recls_sizedict("../datalab/web"))
```

```python
#EX8_12 入力したテキストがnegative傾向が強いかpositive傾向が強いかを判定する関数を作る
positive_words = {"happy": 1, "good": 1, "excellent": 2, "amazing": 2, "great":1, "love": 2, "wonderful": 2}
negative_words = {"sad": -1, "bad": -1, "terrible": -2, "awful": -2, "horrible":-2, "hate": -2, "worst": -2}
negation_words = ["not", "never","no","hardly","barely"]
intensifiers = ["very", "extremely", "quite", "so", "really"]

def get_score(word, positive_words, negative_words, negate): #単語のscoreを調べる関数
  score = 0
  if word in positive_words: #positive表現がある場合
    score = positive_words[word]
  elif word in negative_words:
    score = negative_words[word] #negative表現がある場合
  return -score if negate else score #negateがある(=True)の時、scoreの符号を反転

def apply_modifiers(words, index, negation_words, intensifiers): #単語にnegateとintensifierの有無を調べる関数
  negate = False #初期値False
  modifier = 1 #初期値1倍
  for i in range(max(0, index - 2), index): #この部分は調べて書きました
    preword = words[i]
    if preword in negation_words: #negateがある場合
      negate = True
    elif preword in intensifiers: #intensifierがある場合
      modifier = 2
  return negate, modifier

def analysis_emotion(text): #入力からscoreを取得し、emotionを判定
  total_score = 0
  words = text.split() #textを単語に区切る
  index = 0
  for word in words: #単語ごとにループ、2つの関数に通す
    negate, modifier = apply_modifiers(words, index, negation_words, intensifiers)
    score = get_score(word, positive_words, negative_words, negate)
    score *= modifier
    total_score += score #スコア記録
    index += 1

  if total_score > 0: #スコアからemotionを判定
    emotion = "positive"
  elif total_score < 0:
    emotion = "negative"
  else:
    emotion = "neutral"
  return emotion, total_score

text = input("Enter text to analze: ")
emotion, score = analysis_emotion(text)

print(f"Emotion: {emotion}, Score: {score}")
#大変でした泣
```
