```python
#EX4_1 生徒の名前とスコアを2通りの方法で表示する
scoredict = {'a':94,'b':62,'c':58,'d':40,'e':80 }

for key in scoredict.keys():
  print(key, scoredict[key])
#方法1.dict[key]でスコアを呼び出した

"""
for key,val in scoredict.items():
  print(key,val)
方法2.2つの変数を同時に割り当てた
"""

#EX4_2 60点未満の生徒の名前とスコアのみを表示する
for key,val in scoredict.items():
  if val < 60:
    print("60点未満の生徒:", key, "スコア", val)
#EX3_5で書いたプログラムよりも簡潔になった

#EX4_3 22年卒の生徒数を数える
yeardict = {'a':22,'b':22,'c':23,'d':23,'e':24}

count = 0
for val in yeardict.values():
  if val == 22:
    count += 1
print("22卒の生徒数:", count)
#結果が2になることを確認

#EX4_4_1 22年卒の生徒の平均点を求める
student_22 = []
for key, val in yeardict.items():
  if val == 22:
    student_22.append(key) #yeardictのvalueが22の生徒をstudent_22リストに記録
  total = 0
for i in student_22:
  total += scoredict[i] #student_22リストに格納されたキーを使ってスコアを取得
print("22卒の生徒の平均点:", total/len(student_22))
#平均点が78.0になることを確認

#EX4_4_2 各学年の平均点を求める
ave_students = {}
for year in yeardict:
    key = yeardict[year] #学年をキーとする
    if key not in ave_students:
      ave_students[key] = [] #キーに対応する値としてスコアリストを割り当てる
    ave_students[key].append(scoredict[year]) #その学年の生徒の点数をスコアリストに格納
for k, v in ave_students.items():
  mean = sum(ave_students[k]) / len(ave_students[k])
  print("学年:", k, "平均点:", mean)
#学年ごとに平均点を求めることができた
```

```python
#EX4_5 最もカロリーの高いメニューとカロリーを表示する
kcaldict = {'udon':300, 'salad':150, 'gyudon':500, 'pasta':450}
pricedict = {'udon':200,'gyudon':450, 'salad':150, 'pasta':350 }

maxkcal = 0 #カロリーが最大になるメニューとカロリーの変数を設定
maxmenu = 0
for key, val in kcaldict.items(): #最大値が更新されたら2つの変数を更新
  if val > maxkcal:
    maxkcal = val
    maxmenu = key
print("最もカロリーの高いメニュー:", maxmenu, maxkcal, "kcal")
#最もカロリーが高いメニューが表示されることを確認

#EX4_6 kcal/円が最大になるメニューを表示する
maxkcal_per_yen = 0
maxmenu_name = 0
for key in kcaldict.keys():
  kcal_per_yen = kcaldict[key] / pricedict[key]
  if kcal_per_yen > maxkcal_per_yen:
    maxkcal_per_yen = kcal_per_yen
    maxmenu_name = key
print("kcal/円が最大のメニュー:", maxmenu_name, "1円あたり", maxkcal_per_yen, "kcal")
#kcal/円が最大のメニューがudonになること、udonのkcal/円が正しく表示されることが確認できた

#EX4_7 400円以下のメニューでkcal/円が最大になるメニューを表示する
maxkcal_per_yen = 0
maxmenu_name = 0
for key in kcaldict.keys():
  if pricedict[key] >= 400: #400円以上のメニューを除外する
    break #print(key)で400円未満のudonとsaladだけが表示されることを確認
  else:
    kcal_per_yen = kcaldict[key] / pricedict[key] #EX4_6と同じ処理
  if kcal_per_yen > maxkcal_per_yen:
      maxkcal_per_yen = kcal_per_yen
      maxmenu_name = key
print("400円以下でkcal/円が最大のメニュー:", maxmenu_name, kcaldict[maxmenu_name], "kcal,", pricedict[maxmenu_name], "円")
#400円以下でkcal/円が最大のメニューがudonになること、udonのkcalと値段が正しく表示されることが確認できた

#EX4_8 各メニューの価格とkcalを散布図にする
import matplotlib.pyplot as plt
xlist = []
ylist = []
for key in pricedict.keys():
  xlist.append(pricedict[key]) #価格をx軸とする
  ylist.append(kcaldict[key]) #カロリーをy軸とする
plt.scatter(xlist,ylist, marker='o') #散布図を描画
plt.show()
#見本と同じ散布図が生成できた
```

```python
#EX4_9 サイコロを10回投げた時の平均をグラフにする
import matplotlib.pyplot as plt
import random
freq_list = []
mean_list = []
for i in range(10000):
  numroll = 10
  sumval = 0
  for _ in range(numroll):
    sumval += random.randint(1,6)
  mean = sumval/numroll #平均が約3.5になることを確認
  mean_list.append(mean)
plt.plot(mean_list, marker='o') #棒グラフを描画
plt.show()

plt.hist(mean_list,bins=10) #ヒストグラムを描画
plt.show()
#3.5回〜が一番多くなった
```

```python
#EX4_10 リストの中の文字列の長さによってグループ分けした辞書を作る
words = ["illusion", "spicy", "armageddon", "supernova","drama", "hot mess"]
sorted_words = {}
for word in words:
    key = len(word) #キーとして文字数を割り当てる
    if key not in sorted_words:
        sorted_words[key] = [] #キーに対応する値として空のリストを割り当てる
    sorted_words[key].append(word)
print(sorted_words)
#期待する出力になることを確認
```
