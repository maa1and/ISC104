```python
#p4-5
scoredict = {'a':94,'b':62,'c':58,'d':40,'e':80 } #辞書名 = {key1:value1, key2:value2, ...}
print(scoredict['a']) #辞書名[キー]
scoredict['c'] = 68
scoredict['c'] += 10
scoredict['f'] = 100
print(scoredict)
print(len(scoredict)) #キーと値の数
```

```python
#p6
scoredict = {'a':94,'b':62,'c':58,'d':40,'e':80 }
print(scoredict.keys()) #keyを取り出す
print(scoredict.values()) #値を取り出す
print(scoredict.items()) #ペアで取り出す

for key in scoredict.keys():
  print(key, scoredict[key])
for key in scoredict: #scoredirect.keys()はscoredirectに置き換え可能
  print(key, scoredict[key])
for val in scoredict.values():
  print(val)
for key,val in scoredict.items(): #2つの変数を割りあてられる
  print(key,val)

sum = 0 #全員のスコアの平均を求める
for val in scoredict.values():
  sum += val
print("GPA:",sum/len(scoredict))
```
